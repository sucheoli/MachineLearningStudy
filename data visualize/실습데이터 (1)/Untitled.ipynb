```python
import pandas as pd
df = pd.read_csv('인구동태건수_및_동태율_추이_출생_사망_혼인_이혼__20231031130744.csv', encoding = 'utf-8')
df
```

```python
df = df.set_index('기본항목별').T # df.T.columns vs. df.set_index('기본항목별').T.columns
df
```

```python
from collections import defaultdict
year2value = defaultdict(float)
for y, v in df['출생아수(명)'].items():
    if y[-2] == '7':
        year2value[0] += float(v)/10
    elif y[-2] == '8':
        year2value[1] += float(v)/10
    elif y[-2] == '9':
        year2value[2] += float(v)/10
    elif y[-2] == '0':
        year2value[3] += float(v)/10
    elif y[-2] == '1':
        year2value[4] += float(v)/10
    elif y[-2] == '2':
        year2value[5] += float(v)/3
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

#폰트 수정
plt.rcParams['font.family'] ='Malgun Gothic'
# 윈도우에서는 "C:\Windows\Fonts" 폴더에, 맥에서는 "/Library/Fonts" 폴더에 저장되어 있음
plt.rcParams['axes.unicode_minus'] = False

#축설정
fig, ax = plt.subplots()
xs = df['출생아수(명)'].index.values
ys = df['출생아수(명)'].values.astype(int)
ax.bar(xs, ys)
ax.set_ylabel('출생아수(명)')
ax.set_title('연도별 출생아수')
plt.xticks(rotation=90)
ax.set_xticks(np.arange(0, len(xs)+1, 10))
ax.ticklabel_format(style='plain', axis="y")
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

#색설정
colors = []

cmap = mpl.colormaps.get_cmap('Blues')
cmap = cmap.resampled(10) # Why 10 rather than 6?
for x in xs:
    if x[-2] == '7':
        colors.append(cmap(4))
    elif x[-2] == '8':
        colors.append(cmap(5))
    elif x[-2] == '9':
        colors.append(cmap(6))
    elif x[-2] == '0':
        colors.append(cmap(7))
    elif x[-2] == '1':
        colors.append(cmap(8))
    elif x[-2] == '2':
        colors.append(cmap(9))
ax.bar(xs, ys, color = colors)

bars = ax.bar(xs, ys, color = colors)
poss = np.arange(0, len(xs)+1, 10)
for i, pos in enumerate(poss):
    bar = bars[pos]
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()*0.4, '{:,.0f}'.format(round(year2value[i])),
        bbox = dict(facecolor = 'white', alpha =.8))
    
#legend설정
ax.legend(labels = [xs[pos] for pos in poss], handles =[bars[pos] for pos in poss], title='년대')
plt.show()
```

```python
from matplotlib.ticker import PercentFormatter

fig, axs = plt.subplots(1, 2, tight_layout=True, figsize=(8, 4))
xs = df['중간고사'].values
axs[0].set_ylabel('학생수')
axs[0].set_title('중간고사 점수분포(학생수)')
axs[1].set_title('중간고사 점수분포(퍼센티지)')

cmap = mpl.colormaps.get_cmap('Blues')
cmap = cmap.resampled(15)
colors = [cmap(i+5) for i in range(10)]

n, bins, patches = axs[0].hist(xs, bins = 10, rwidth = 0.95)
axs[0].set_xticks(bins)
axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=90)
axs[0].grid(color = 'gray', linewidth = 0.2, axis = 'y')
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)
n, bins, patches = axs[1].hist(xs, bins = 10, density = True, rwidth = 0.95)
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
axs[1].set_xticks(bins)
axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=90)
axs[1].grid(color = 'gray', linewidth = 0.2, axis = 'y')
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.show()
```
